```python
# câu 8:  xếp chồng 2 mảng theo chiều dọc

import numpy as np

# tạo data 2 mảng 2d (2 hàng 5 cột)
# tạo mảng hai chiều (2 hàng 5 cột) từ 0 đến 9
arr1 = np.arange(10).reshape(2,-1)
print('arr1: \n', arr1)
# tạo mảng hai chiều (2 hàng 5 cột) chỉ có giá trị 1
arr2 = np.ones((2, 5), dtype=np.int32)
print('arr2: \n',arr2)

# Cách 1 :
c = np.vstack((arr1, arr2))
print("Cách 1: \n", c)
print(c.shape)

# Cách 2 :
d = np.concatenate((arr1, arr2), axis=0)
print("Cách 2: \n", d)
print(d.shape)

# Cách 3 :
e = np.stack((arr1, arr2), axis=0)
print("Cách 3 \n",e)
print(e.shape)

f = np.r_[arr1, arr2]
print(f)
print(f.shape)
```

```python
# câu 15a : Tạo hàm xử lý trên mảng numpy

import numpy as np

def maxx(x,y):
    return x if x > y else y


a = np.array([5,7,9,8,6,4,5])
b = np.array([6,3,4,8,9,7,1])

# So sánh các cặp phần tử trong mảng a và b (theo index) sau đó lấy ra giá lớn hơn trong từng cặp

# dùng numpy.vectorize lấy hàm maxx: x,y và biến nó thành maxx: a[] -> b[].
pair_max = np.vectorize(maxx)
print(pair_max(a,b))
```

```python
# câu 22: Lấy vị trí của n giá trị đứng đầu từ một mảng numpy
import numpy as np

# cho mảng a chứa các số ngẫu nhiên từ [1, 50)
a = np.random.randint(1, 50, (5, 3))
print("Mảng a: \n", a)

# Flatten mảng để dễ lấy vị trí
flat = a.ravel()   # hoặc a.flatten()

# -------------------------
# Cách 1: Dùng np.argsort()
# -------------------------
print("\nCách 1 (argsort)")
sorted = np.argsort(flat)
idx = sorted[-5:]           # lấy 5 index lớn nhất
print("Vị trí :", idx)
print("Giá trị :", flat[idx])


# -------------------------
# Cách 2: Dùng np.argpartition(), giới thiệu thêm cho các bạn 1 cách khác ngoài argsort
# -------------------------
print("\nCách 2 (argpartition)")
idx2 = np.argpartition(flat, -5)[-5:]  # lấy 5 vị trí lớn nhất
print("Vị trí :", idx2)
print("Giá trị :", flat[idx2])

# Lưu ý kết quả là ngẫu nhiên nên kết quả có thể thay đổi mỗi khi chạy
```

```python
import numpy as np

# Ma trận A và B
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[2, 0],
              [1, 2]])

# Tính AB và BA
AB = A @ B
BA = B @ A

print("A =\n", A)
print("B =\n", B)
print("AB =\n", AB)
print("BA =\n", BA)

print("AB == BA  -->", np.all(AB == BA))
```

```python
import numpy as np

# Tạo hai ma trận A và B tùy ý
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8, 9],
              [1, 2, 3]])

print("A =\n", A)
print("B =\n", B)

# Tính (A + B)^T
left = (A + B).T

# Tính A^T + B^T
right = A.T + B.T

print("\n(A + B)^T =\n", left)
print("\nA^T + B^T =\n", right)

print("\nHai ma trận bằng nhau? ->", np.all(left == right))
```

```python
import numpy as np

# Ma trận A kích thước 2×3
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Ma trận B kích thước 3×2
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print("A =\n", A)
print("B =\n", B)

# Tính AB
AB = A @ B
print("\nAB =\n", AB)

# Tính (AB)^T
left = AB.T
print("\n(AB)^T =\n", left)

# Tính B^T A^T
right = B.T @ A.T
print("\nB^T A^T =\n", right)

print("\nHai kết quả bằng nhau? ->", np.all(left == right))

```
